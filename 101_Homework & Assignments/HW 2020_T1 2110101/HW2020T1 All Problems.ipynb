### Task HW_2020_01_02 : Parametric Equations

```python
# Prog-02: Beautiful Parametric Equation 
# 6730084521 Chatrphol Ovanonchai

# 1 : write LaTex equations (with $ ends)
# $ x(t) = sin(t) * e^( cos(t)) - 2cos(4t) - sin(t/12)^5 ) $
# $ y(t) = cos(t) * e^( cos(t)) - 2cos(4t) - sin(t/12)^5 ) $

# This program generates Butterfly Eqautions
# I've chosen this equation by myself. 
 
# 2 : import math & matplotlib
import numpy as np
import matplotlib.pyplot as plt 

# ------------- 
# 3 : setting up period (T)
def setup_period(min_t, max_t, dt): # range = min_t & max_t 
    # store period list (T) & append each value
    T = []; t = min_t
    while t <= max_t:              
        T.append(t) 
        t += dt 
    if t != max_t: T.append(max_t) 
    return T 
#------------------------------------ 
# 4 : plotting x,y points
def plot(x, y, min_t, max_t, dt): 
    T = setup_period(min_t, max_t, dt) 
    X = [x(t) for t in T]  # plot each point in x-axis
    Y = [y(t) for t in T]  # plot each point in y-axis
    plt.plot(X, Y, 'pink') 

# ==================================== 
# 5 : define x,y
def x_eq(t): 
    xt = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12) ** 5)
    return xt 
 
def y_eq(t): 
    yt = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12) ** 5)  
    return yt 
 
# 6 : call fn plot (x&y) & show graph
plt.title('Parametric Equations 6730084521')
plot(x_eq , y_eq , -100, 100, 0.01) # call setup_period fn
```

```python
## Extra : Moo-Deng Equations (Modded By Me & Generated By ChatGPT)
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import binom

# Function to plot a parametric ellipse
def parametric_ellipse(h, k, a, b, t_range, color='black', label=None):
    t = np.linspace(t_range[0], t_range[1], 100)
    x = h + a * np.cos(t)
    y = k + b * np.sin(t)
    plt.plot(x, y, color=color, label=label)

# Function to plot a circle
def plot_circle(h, k, r, color='black', label=None):
    t = np.linspace(0, 2*np.pi, 100)
    x = h + r * np.cos(t)
    y = k + r * np.sin(t)
    plt.plot(x, y, color=color, label=label)

# Function to compute Bezier curve points
def bezier_curve(P, t):
    n = len(P) - 1
    x, y = 0, 0
    for i in range(n + 1):
        x += binom(n, i) * (1 - t)**(n - i) * t**i * P[i][0]
        y += binom(n, i) * (1 - t)**(n - i) * t**i * P[i][1]
    return x, y

# Function to plot Bezier curve
def plot_bezier(P, color='black'):
    t_vals = np.linspace(0, 1, 100)
    curve = [bezier_curve(P, t) for t in t_vals]
    x_vals, y_vals = zip(*curve)
    plt.plot(x_vals, y_vals, color=color)

# Set up the plot
plt.figure(figsize=(6, 6))
plt.axis('equal')

# Head (Ellipse)
parametric_ellipse(h=0, k=0, a=3, b=2.2, t_range=[0, 2*np.pi], color='gray')

# Eyes (Circles)
plot_circle(h=-1, k=0.9, r=0.3, color='black')
plot_circle(h=1, k=0.9, r=0.3, color='black')

# Nostrils (Small Circles)
plot_circle(h=-0.6, k=0.1, r=0.2, color='grey')
plot_circle(h=0.6, k=0.1, r=0.2, color='grey')

# Ears (Parametric Ellipses)
parametric_ellipse(h=-2.3, k=1.7, a=0.5, b=0.7, t_range=[0, 2*np.pi], color='pink')
parametric_ellipse(h=2.3, k=1.7, a=0.5, b=0.7, t_range=[0, 2*np.pi], color='pink')

# Mouth (Bezier Curve)
P_mouth = [(-1.5, -0.8), (-0.8, -1.4), (0.8, -1.4), (1.5, -0.8)]
plot_bezier(P_mouth, color='pink')

# Add labels and title
plt.title("Simplified Moo-Deng 6730084521")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show plot
# plt.legend() ==> for showing label
plt.show()
```

### Task HW_2020_01_04 : Guessing Number Conditions

```python
76# Case-Scenario 1 : Guessing Number by Input Number (Define Start-End By Myself) (Using Bisection Idea)

# 1 : define start & end & input guess number
start = int(input('Enter Start Guessing Number Range : '))
end = int(input('Enter End Guessing Number Range : '))
winner = False # check if guess right number

# 2 : structure of program
import random
real_num = random.randint(start , end)

round_left = 10 
print(f"Guessing Number from {start} to {end}")

for round_left in range(9,-1,-1) : # rp = round passed (start from 9 to 0)
    # Enter Guess Number
    guess_num = int(input('Guessing Number : '))
    
    # case 1 : guess_num > real_num : tell user for lower number (dont forget to change start & end number)
    if guess_num > real_num : 
        print(f"{guess_num} is higher than real number , Input Lower Number")
        
        # change end (if new guess is less than end)
        if guess_num < end : 
            end = guess_num - 1
    
    # case 2 : guess_num < real_num : tell user for higher number (dont forget to change start & end number)
    elif guess_num < real_num : 
        print(f"{guess_num} is lower than real number , Input Higher Number")
        
        # change start (if new guess is more than start)
        if guess_num > start : 
            start = guess_num + 1
        
    # case 3 : same number -> break
    elif guess_num == real_num :
        winner = True
        print(f'Congratulations , You Guesses Right . The Real Number is {real_num}')
        break
    
    # still input new number
    print(f"Round Left #{round_left} : Guessing Number from {start} to {end}")
    
# check winner
if winner == False : 
    print('Bad Luck , Try Again Next Time . Real Number is' , real_num)
```
