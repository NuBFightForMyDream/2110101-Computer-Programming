โดยที่

• `𝑥[𝑖], 𝑦[𝑖]` แทนพิกัดคาร์ทีเซียนบนระนาบสองมิติของนกตัวที่ i

• `𝑑𝑥[𝑖] , 𝑑𝑦[𝑖]` แทนทิศทางของนกตัวที่ i เคลื่อนที่ไป โดย
`(𝑑𝑥[𝑖], 𝑑𝑦[𝑖])` เป็น 𝑢𝑛𝑖𝑡 𝑣𝑒𝑐𝑡𝑜𝑟 คือ

`(𝑑𝑥[𝑖])² + (𝑑𝑦[𝑖])² เท่ากับ 1` (อาจแทนทิศทางด้วยมุมในช่วง `[0, 2π)` ก็ได้ แต่การบ้านนี้เลือกแทนทิศทางด้วยเวกเตอร์คาร์ทีเซียน)

`𝑥 = [5, 10, 15, 20]` , `𝑦 = [10, 5, 10, 5]` , `dx = [0,1/√2,0,-1]` , `dy = [1,1/√2,-1,0]`

1. **`def gen_data(N, W, H):`**

    -> **`W, H`** เป็นจํานวนจริงเก็บความกว้างและความสูงของวินโดว์ที่แสดงกลุ่มนก

    => `return` : ลิสต์ 4 ตัว **`x, y, dx และ dy`** ที่เก็บข้อมูลตําแหน่งและทิศทางของนก `N` ตัวที่สุ่มสร้างขึ้นมา

    => **`d`** เป็นจํานวนจริงแทนการกระจัด (displacement) ที่นกจะเปลี่ยนตําแหน่งไปในทิศทาง `dx, dy`

    => **`W, H`** เป็นจํานวนจริงแทนความกว้างและความสูงของวินโตว์ที่แสดงกลุ่มนก
