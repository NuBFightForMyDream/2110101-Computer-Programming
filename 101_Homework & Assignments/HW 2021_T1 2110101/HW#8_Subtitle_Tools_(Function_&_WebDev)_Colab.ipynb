## รูปเเบบของเเฟ้ม .srt

ซอฟต์แวร์แสดงภาพยนตร์ทั้งหลาย จะให้ผู้ชมเลือกแฟ้มบทบรรยายได้ รูปแบบบทบรรยายใต้ภาพที่ได้รับความนิยมรูปแบบหนึ่งคือ srt (SubRip
subTitle https://en.wikipedia.org/wiki/SubRip) เป็นแฟ้มข้อความ ภายในประกอบด้วยบทบรรยายหลาย ๆ ข้อความ เรียงตามเวลาที่เริ่มและ
เลิกแสดงข้อความ แต่ละบท มีอย่างน้อย 4 บรรทัด ดังนี้

- เลขลําดับของบทบรรยาย

- เวลาเริ่มแสดง --> เวลาเลิกแสดง
เวลาอยู่ในรูปแบบ HH:MM:SS,ms แทนเลขชั่วโมง เลขนาที เลขวินาที และเลขมิลลิวินาที

(โดย 3 จํานวนแรกเป็นเลขสองหลัก ส่วนหลักที่ 4 คือ เลขมิลลิวินาที ที่เป็นเลขสามหลัก เช่น 00:12:02,103)

- ข้อความบรรยาย มีได้หลายบรรทัด โดยบรรทัดสุดท้ายต้องเป็นบรรทัดว่าง)

เช่น ข้อความในตารางข้างล่างนี้ ทางซ้ายแสดงบทบรรยายปกติทั่วไป ส่วนทางขวาเป็นแบบ SDH และยังมีการใส่สีกํากับข้อความต่าง ๆ ด้วย

## **Step 1 : ฟังก์ชั่นที่ต้องเขียน**

1. **`def shift(file_in, time_shift, file_out) :`**
  
    => **`file_in`** เป็นสตริงระบุชื่อแฟ้ม .srt (แฟ้มมี `encoding` เป็น `utf-8`)
  
    => **`time_shift`** เป็นจํานวนเต็มแทนปริมาณมิลลิวินาที (เป็นได้ทั้งค่าบวก ศูนย์และ ค่าลบ)

    => **`file_out`** เป็นสตริงระบุชื่อแฟ้มผลลัพธ์ที่มีการปรับเวลาบทบรรยายของ `file_in` (แฟ้มมี `encoding` เป็น `utf-8`)

    => ฟังก์ชันนี้ไม่ **return** อะไร สิ่งที่ฟังก์ชันทํา คืออ่านข้อมูลบทบรรยายจากแฟ้ม `file_in` แล้วบวกค่าเวลาทั้งหลายไปอีก `time_shift`
      
      - ถ้า `time_shift` เป็นค่าบวก แสดงว่า ต้องการให้แสดงบทบรรยายช้าลงจากเดิม
      - ถ้า `time_shift` เป็นค่าลบ แสดงว่า ต้องการให้แสดงบทบรรยายเร็วขึ้นจากเดิม
        - ถ้าบวกเวลาด้วยค่าลบแล้ว ได้เวลาเป็นค่าลบ ให้แทนเวลาด้วยค่า **00:00:00,000**
        - ถ้าบวกเวลาด้วยค่าลบแล้ว ได้เวลาทั้งเริ่มและเลิก น้อยกว่าหรือเท่ากับศูนย์ทั้งคู่ ให้ลบบทนั้นทิ้ง (ดูตัวอย่างข้างล่าง)
    
  => **ตัวอย่าง** : แฟ้ม test.srt มีข้อมูลดังแสดงข้างล่างนี้ทางซ้าย

    คําสั่ง `shift('test.srt', 4500, 'test1_shifted.srt')`

2. **`def merge(base_file, merge_file, threshold, file_out):`**
  
    => **`base_file`** เป็นสตริงระบุชื่อแฟ้ม .srt (แฟ้มมี `encoding` เป็น `utf-8`)
  
    => **`merge_file`** เป็นสตริงระบุชื่อแฟ้ม srt (แฟ้มมี `encoding` เป็น `utf-8`)

    => **`threshold`** เป็นจํานวนเต็ม (ไม่ติดลบ)
    
    => **`file_out`** เป็นสตริงระบุชื่อแฟ้มผลลัพธ์จากการผสาน `base_file` และ `merge_file` (แฟ้มมี `encoding` เป็น `utf-8`)

    => ฟังก์ชันนี้ไม่ **return** อะไร ทําหน้าที่ผสานบทบรรยายในแฟ้ม `base_file` และ `merge_file` เข้าด้วยกัน โดยมีหลักการผสานดังนี้`time_shift`

      - นําแต่ละบทบรรยายใน `merge_file` ไปรวมกับบทบรรยายใน `base_file` **ที่มีเวลาเริ่มใกล้เคียงกันที่สุดและต้องใกล้กันไม่เกิน
ค่า threshold** หลังรวมแล้วให้เลือกใช้เวลาเริ่มและเลิกแสดงของ `base_file` ในแฟ้มผลลัพธ์

        เช่น จากตัวอย่างข้างล่าง บทบรรยายที่มีสีพื้นเหมือนกัน ถูกรวมเข้าด้วยกัน ในที่นี้กําหนดให้ `threshold` มีค่า 1000
      - บทบรรยายที่ไม่มีสีพื้นครอบในตัวอย่าง คือบทที่ไม่สามารถหาบทบรรยายของอีกแฟ้มที่ใกล้เคียงกันที่ไม่เกินค่า `threshold` ได้

3. **`def clean(file_in, file_out):`**
  
    => **`file_in`** เป็นสตริงระบุชื่อแฟ้ม srt (แฟ้มมี `encoding` เป็น `utf-8`)
  
    => **`file_out`** เป็นสตริงระบุชื่อแฟ้มผลลัพธ์ (แฟ้มมี `encoding` เป็น `utf-8`)

    => ฟังก์ชันนี้ไม่ **return** อะไร ฟังก์ชันนี้ไม่คืนอะไร สิ่งที่ฟังก์ชันทํา คือ อ่านข้อมูลบทบรรยายจากแฟ้ม `file_in` เพือสร้างแฟ้ม `file_out` ตามขั้นตอนข้างล่างนี้

      - 1. ลบข้อความใน file_in ที่อยู่ระหว่างเครื่องหมาย `( กับ ) , [ กับ ] , { กับ } และ < กับ >`
`(รวมทั้งเครื่องหมาย ( ), [ ], { } และ < > )` ออกให้หมด

          คู่เครื่องหมายเหล่านี้ไม่จําเป็นต้องอยู่ในบรรทัดเดียวกัน
และไม่มีกรณีที่เครื่องหมายเหล่านี้ปรากฏซ้อน ๆ กัน เช่น `<[b<b>]>`

      - 2. หลังทําขั้นตอนที่ 1 ให้ลบบรรทัดของบทที่ข้อความบรรยาย**ไม่มีตัวอักษรและตัวเลข**เลย (อาจมีเครื่องหมายอื่นอยู่) ออก

  การทดสอบว่า ตัวอักขระในตัวแปร `c` เป็นตัวอักษรหรือตัวเลขหรือไม่ ให้ใช้ `c.isalnum()` ซึ่งคืน `True` หรือ `False`
(ห้ามทดสอบด้วยวิธีอื่น)

      - 3. หลังทําขั้นตอนที่ 2 หากบทบรรยายไม่มีคําบรรยายเหลืออยู่เลย ให้ลบบทนั้นออก

      - 4. บันทึกบทบรรยายที่เหลือไปยังแฟ้ม `file_out`

Example :
คําสั่ง `clean('test.srt', 'test_cleaned.srt')`

สร้างแฟ้ม `'test_cleaned.srt'` ที่มีข้อมูลดังแสดงข้างล่างนี้ทางขวา

- `test2_en_cleaned.srt, test2_th_cleaned.srt, test3_en_cleaned.srt และ test3_th_cleaned.srt`

เป็นแฟ้มที่ได้จากการ `clean` แฟ้ม `test2_en.srt, test2_th.srt, test3_en.srt และ test3_th.srt` ตามลําดับ

- `test3_en_th_merged.srt` เป็นแฟ้มบทบรรยายภาษาอังกฤษและไทย ที่ได้มาจากใช้คําสั่ง
`merge('test3_en.srt', 'test3_th.srt', 1000, 'test3_en_th_merged.srt')`

- `test3_th_en_merged.srt` เป็นแฟ้มบทบรรยายภาษาอังกฤษและไทย ที่ได้มาจากใช้คําสั่ง
`merge('test3_th.srt', 'test3_en.srt', 1000, 'test3_th_en_merged.srt')`

- `test4_en_th_merged.srt` เป็นแฟ้มที่ได้จากคําสั่ง
`merge('test4_en.srt', 'test4_th.srt', 1000, 'test4_en_th_merged.srt')`

- แฟ้ม `srt` บางแฟ้ม อาจมีค่าของเลขลําดับบท ไม่เป็นไปตามกฎ คือ ไม่ได้เรียง `1,2,3, ...` ดังนั้น ชุดคําสั่งที่อ่านเลขลําดับไม่ต้องสนใจ
ค่าของเลขลําดับในแฟ้มที่อ่าน

  แต่แฟ้มผลลัพธ์ที่สร้างจากฟังก์ชันที่ให้เขียนในการบ้านนี้ ต้องมีเลขลําดับที่เรียง `1,2,3,...` ตามข้อกําหนด

- เนื่องจากแฟ้ม `srt` ถูกบันทึกไว้โดยเข้ารหัสในรูปแบบ `utf-8` ดังนั้น คําสั่งที่เปิดแฟ้มเพื่ออ่าน หรือเพื่อบันทึก จะต้องมี `encoding='utf-8'` อยู่ในคําสั่ง `open` ด้วย เช่น
`f_in = open(file_in , encoding='utf-8')`
หรือ
`f_out = open(file_out, 'w', encoding='utf-8')`
