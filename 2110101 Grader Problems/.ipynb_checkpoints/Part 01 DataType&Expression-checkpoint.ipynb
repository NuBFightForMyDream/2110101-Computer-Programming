### Task 01_01 : Stirling Factorial (1 Star)


```python
# Logic : รับค่า n! -> เข้าสูตร ขอบบน-ล่าง ! -> print ค่าขอบออกมา
# Step : 
# 1. รับ input n (int)
# 2. กำหนด ตป. n_min , n_max (เข้าสูตร)
# 3. print ขอบบน , ขอบล่าง

# Input : รับ input n (int) + import math
import math # เพื่อใช้ฟังก์ชั่นทางคณิตศาสตร์
n = int(input()) #number for n! (int)

# Process : กำนหดค่า n_min , n_max
n_floor = math.sqrt(2*math.pi) * n**(n+0.5) * math.exp(-n + 1/((12*n)+1))
n_ceil = math.sqrt(2*math.pi) * n**(n+0.5) * math.exp(-n + 1/(12*n))

# Output : print ขอบบน , ขอบล่าง ของ n!
print(n_floor) #float
print(n_ceil) #float

## อย่าลืม!! เวลาเอาไปลงใน Grader -> ลบ command ใน input() ออกด้วย

# Test Case : 
# case 1 : n = 1 -> 1! -> n_floor = 0.995870161462797 , n_ceil = 1.0022744491822266
# case 2 : n = 5 -> 5! -> n_floor = 119.96985395920886 , n_ceil = 120.00263708619694
# case 3 : n = 50 -> 50! -> n_floor = 3.041400953459947e+64 , n_ceil = 3.041409387750485e+64
# case 4 : n = 100 -> 100! -> n_floor = 9.332615094728949e+157 , n_ceil = 9.332621570317617e+157
```

### Task 01_02 : Quadratic Root (1 Star)


```python
# Step : 
# 1. รับค่า a,b,c เพื่อเข้ารูปสมการ ax^2+bx+c = 0
# 2. เข้าสูตรเพื่อหาค่า x -> มี x 2 คำตอบ
# 3. print ค่า x และ round ค่า

# Input : a , b , c (float) , import math module
import math
a = float(input()) #x^2 degree #float
b = float(input()) #x^1 degree #float
c = float(input()) #x^0 degree #float

# Process : เข้าสูตร quadratic
x_1 = (-b - math.sqrt( (b**2) - (4*a*c) )) / (2*a)
x_2 = (-b + math.sqrt( (b**2) - (4*a*c) )) / (2*a)

# Output : print & round !! บรรทัดเดียวกัน อย่าลืม !!
print( round(x_1 , 3) , round(x_2 , 3) ) #ค่าซ้ายเป็น x1 ค่าขวาเป็น x2

## ถ้า b^2 - 4ac = 0 -> ระบบจะบอกว่า math domain error
## วิธีเเก้ -> เราจะทำการเช็ค b^2 - 4ac (discriminant) ก่อน
## ถ้า b^2 - 4ac มากกว่า หรือ เท่ากับ 0 -> ให้ output x1 , x2 ออกมา
## ถ้า b^2 - 4ac น้อยกว่า 0 -> บอก user ว่า "Error"

# test case (case เดิม)
# case 1 : a = 1 , b = -5 , c = 6 -> x^2 -5x + 6 = 0        -> x_1 = 2 , x_2 = 3 (x_1 < x_2)
# case 2 ; a = 1 , b = -1 , c = 42 -> x^2 -x -42 =  0       -> x_1 = -6 , x_2 = 7 (x_1 < x_2)
# case 3 ; a = 6 , b = -4 , c = -12 -> 6x^2 -4x -12 =  0    -> x_1 = -1.12 , x_2 = 1.786 (x_1 < x_2)
# case 4 ; a = 20 , b = -50.5 , c = -21.2 -> 20x^2 -50.5x -21.2 =  0 -> x_1 = -0.367 , x_2 = 2.892 (x_1 < x_2)
```

```python
## ปรับปรุง code : 01_02 Quadratic Roots
# Step 1 : รับ input a,b,c (float) แทนดีกรี x^2 , x^1 , x^0 ตามลำดับ + import math
import math
a = float(input()) #x^2 degree (float)
b = float(input()) #x^1 degree (float)
c = float(input()) #x^0 degree (float)

# Step 2 : check discriminant first
if (b**2) - (4*a*c) >= 0 : 
    # 2.1 : define x_1 , x_2
    x_1 = (-b - math.sqrt( (b**2) - (4*a*c) )) / (2*a)
    x_2 = (-b + math.sqrt( (b**2) - (4*a*c) )) / (2*a)
    # 2.2 : print & round x_1 , x_2
    print( round(x_1 , 3) , round(x_2 , 3) )
    
else : # 2.3 : b^2 - 4ac < 0 -> complex nb. answer
    print('No Real Number Answers, Error 404')
    
# test case :
# case 1 : a = 1 , b = 5 , c = 6 -> x^2 + 5x + 6 = 0 -> x1 = -2 , x2 = 3 
# case 2 : a = 4 , b = 6 , c = 8 -> 4x^2 + 6x + 8 = 0 -> No Real Number
```

### Task 01_03 : Strange Expression (2 Star)


```python
# Step : 
# 1. เเทนสูตรลง a -> เข้าสูตรดีๆ ระวังผิดพลาด ( เรื่อง math fn กับ () )
# 2. round ค่าออกมา

# Step 1 : เเทนลง ตป. & เข้าสูตร !!!!!อย่าลืม import math ด้วยยยย!!!!
# จากรูปสมการ เราจะทำให้อยู่ในรูป answer = (a + b) / c เพื่อกันสับสน\
import math
a = math.pi - ( math.factorial(10) / (8**8) ) 
b = math.log(9.7) ** ( (7/math.sqrt(71) - math.sin(40/180 * math.pi)) ) # ส่วน 'เศษ'
c = 1.2 ** (2.3 ** (1/3)) # ส่วน "ส่วน"
answer = (a + b) / c

# ระวัง 1 : ใช้ 10! ไม่ได้ ต้องใช้เป็น math.factorial(10)
# ระวัง 2 : log ปกติใน python ถ้าไม่ใส่ฐาน จะใช้ฐาน e -> คำสั่ง : math.log(a,base)
# ระวัง 3 : ใน math.sin() มุมด้านในจะเป็น rad -> แปลง deg เป็น rad ก่อน -> 40 องศา = 2pi/9 (pi = 180 องศา) หรือจะใช้ math.radians() ก็ได้

# Step 2 : print & round ค่าตัวเเปร
print( round(answer , 6) )

## เวลาเอาลง grader -> ใส่คำสั่งลงบรรทัดเดียวกัน
```

### Task 01_04 : Body Surface Area (2 Star)


```python
# Logic : เเปลง W,H ให้เข้าสูตร เพื่อหา Area (m^2)
# Step :
# 1. รับ input W (float), H_cm (float) เข้ามา 
# 2. เขาสูตรการหา Area เเต่ละคน !! ระวัง เครื่องหมาย & fn ทางคณิตศาสตร์ ดีๆ !!
# 3. print area ของเเต่ละคนออกมา

# Input : W (float) , H (float) & import math ด้วย เผื่อใช้ fn ทางคณิตศาสตร์
import math
W = float(input()) #input weight (kg)
H = float(input()) #input height (cm)

# Process : เข้าสูตรคำนวณ Area
area_mosteller = math.sqrt(W*H) / 60
area_haycock = 0.024265 * (W**0.5378) * (H**0.3964)
area_boyd = 0.0333 * ( W ** ( 0.6157- (0.0188 * math.log(W,10)) ) ) * (H**0.3)

# ระวัง 1 : ยกกำลัง ใช้เครื่องหมาย ** อย่าไปใช้ ^ เพราะให้ค่าทางคณิตศาสตร์ต่างกัน
# ระวัง 2 : ฟังก์ชั่น log ปกติเป็นฐาน e เราเลยต้องใส่ฐานให้ด้วย ดัง command math.log(a,base)

# Output : print ค่า area ของเเต่ละคนออกมา
print(area_mosteller) #m^2
print(area_haycock) #m^2
print(area_boyd) #m^2
```

### Task 01_05 : Duration (3 Star)


```python
# Logic : หา ช่วงเวลาที่ต่างกัน (เเปลงเป็น sec ก่อน เเล้วเเปลงเป็น  ้พ)
# Step : 
# 1. รับค่า h1,m1,s1 เเละ h2,m2,s2 (int)  
# 2. เเปลงค่า h,m,s ให้เป็น t1 , t2 ตามลำดับ
# 3. define ตป. dt (difference time) = t2 - t1
# 4. เเปลง dt ให้เป็น dh,dm,ds (หารปัดเศษ)
# 5. เเปลงค่า dh จาก - ให้เป็น + => (24 - dh)

# Step 1 : รับ input h1,m1,s1,h2,m2,s2 (int)
h1 = int(input()) 
m1 = int(input())
s1 = int(input())

h2 = int(input())
m2 = int(input())
s2 = int(input())

# Step 2 : เเปลงค่า h,m,s เป็น t 
t1 = h1*60*60 + m1*60 + s1 #time total (sec) initial
t2 = h2*60*60 + m2*60 + s2 #time total (sec) final

# Step 3 : define dt (t_final - t_initial)
dt = t2 - t1

# Step 4 : แปลง dt ให้เป็น dh , dm , ds
dh = dt // (60*60) #หารปัดเศษ 3600 sec -> เอา จน.hr
dt = dt - (dh * 3600) #ลบ จน.ชั่วโมงออก เหลือเศษ hr (sec) ## ตอนนี้ dt เหลือเศษ hr เเล้ว -> เเปลงต่อเป็น m,s

dm = dt // 60 #หารปัดเศษ 60 sec -> เอา จน. min
dt = dt - (dm * 60) #ลบ จน.นาทีออก เหลือเศษ min (sec) ## ตอนนี้ dt เหลือเศษ min แล้ว-> ไม่ต้องเเปลงต่อเเล้ว เพราะที่เหลือเป็นหน่วย sec

ds = dt #เศษ min = จน.sec ที่เหลือ กำหนดให้เป็น ds ได้เลย

# Step 5 : แปลง dt
dh = (s2 - s1) % 24

print( str(dh) + ':' + str(dm) + ':' + str(ds) ) #เขียนใน format HH : MM : SS


## ลองเอาไป run ใน grader เเล้ว

# Note : Sol 1 : ลองใช้ dt = abs(t2 - t1) ได้เเค่ 40 pts เเสดงว่า ยังมีปัญหาอยู่ในบาง testcases -> ลองใช้ mod ดู (น้ำเอกเเนะนำ)
#        Sol 2 : เจอปัญหาว่า dt เป็นลบอยู่ -> เเก้โดย + 24 เข้าไป -> dh = 24 + (h2-h1) -> ได้ 50 คะเเนน -> ยังมีปัญหาอยู่
#        Sol 3 : เจอปัญหาว่า dt เกิน 25 ชม.ได้ -> เเก้โดย mod 24 เพื่อให้ไม่เกิน 24 -> ได้ 70 คะเเนน -> เดะเช็คบาง testcases ที่ผิด 
#        Sol 4 : ตอนนี้เจอปัญหาว่า dh เกินมา 1 ตลอด (ถ้า 23 -> โชว์ 0 , ถ้า 1 -> โชว์ 2) => เเก้โดย
```
