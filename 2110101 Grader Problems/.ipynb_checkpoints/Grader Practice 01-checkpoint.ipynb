### Task P1_01 : Flowchart #01 (2 Star)

```python
# 1 : รับ input 5 ตัว (int) ด้วย split (list comprehension)
x1,x2,x3,x4,x5 = [int(e) for e in input('Enter 5 Values : ').split()] # [x1,x2,x3,x4,x5] = list (of int)

# 2 : ถอดเเต่ละ cd จากโจทย์
# cd 1 : x1-x5 > x2
if (x1 - x5) > x2 : 
    # Yes => cd 1.1 : x2 > x3+x1
    if x2 > (x3 + x1) :
        if (x3 + x5) > x4 :
            print('C5')
        else :
            if x3 < x5 : #ย่อรูป else:if => elif
                print('C6')
            else : 
                print('C7')
    else :
        print('C8') # บังคับ print 

    # No => cd : 1.2 : x2 - x1 > x3
    if not ((x2 - x1) > x3) : #ย่อรูป else:if => elif
        if x4 < (x5 + x1) :
            if (x3 + x2) > x5 :
                print('C3')
            else : 
                print('C2')
            
        else : 
            print('C1')
    else : 
        print('C4') # บังคับ print C4
```
