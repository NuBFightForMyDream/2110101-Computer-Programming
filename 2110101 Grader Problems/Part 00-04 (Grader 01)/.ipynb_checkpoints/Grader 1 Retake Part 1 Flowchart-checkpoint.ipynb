### Task 03_11 : Next 15 Days (2 Star) => 10.25 min


```python
# 1 : input d,m,y + declare normal month + year_ad
d,m,y = [int(e) for e in input().split()]
y = y - 543
n = 31 # day in month

# 2 : check cd.
if not((m == 4) or (m == 6) or (m == 9) or (m == 11)) :
    # do inside
    if m == 2 : # check february
        n = 28
        # 2.1 : inside n == 2-> check leap year
        if y % 400 == 0 :
            n = 29
        if (y % 4 == 0) and (y % 100 != 0) :
            n = 29
        # 2.2 : if not in cd. -> n = 28

    # 2.3 : if m ends with 'kom' -> n = 30
elif ( m==4 or m==6 or m==9 or m==11 ) :
    n = 30

# 3 : plus 15 days
d = d + 15
if d > n : # day > 31 -> mod & +1 month
    d = d - n 
    m = m + 1
    
if m > 12 : # month > 12 -> mod & +1 year
    m = m - 12
    y = y + 1

# change year_ad to year_bc
y = y + 543

# 4 : print format DD/MM/YYYY
print( str(d) + '/' + str(m) + '/' + str(y) )
```
